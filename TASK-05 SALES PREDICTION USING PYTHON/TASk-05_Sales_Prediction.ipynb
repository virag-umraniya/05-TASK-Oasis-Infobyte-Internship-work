## Model Creation

# Step 1 - Independent and dependent variable creation

```python

X = data[['TV', 'Radio', 'Newspaper']] # X variable contains all input
y = data.Sales # Y contains output / Target
```

```python
X
```

```python
y
```

## Model Selection

# Step 2 - Training and testing data creation

```python

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train,y_test = train_test_split (X, y, test_size = 0.2, random_state = 33)
```

```python
X_train
```

```python
X_test
```

```python
X_test.shape
```

```python
y_train
```

```python
y_test
```

```python
y_test.shape
```

# step 3 Model Selection

```python
from sklearn.linear_model import LinearRegression
```

```python
LR = LinearRegression() # Object Creation
```

```python
LR.fit(X_train, y_train) # Training of linear regression
```

```python
y_predict = LR.predict(X_test)
```

```python
y_test # actual values
```

```python
y_predict # Values predicted by model
```

## Step 4 - Model Evaluation

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error, root_mean_squared_error
```

```python
mean_squared_error(y_test, y_predict)
```

```python
r2_score(y_test, y_predict)
```

```python
# calculation of adjusted r2 score
adjusted_r2 = 1-(1-0.90) * (40 - 1) / (40 - 3 - 1)
# adj_r2 = 1-[(1-r2)(n-1)/(n-k-1)] 
# where k = number of independent feature
adjusted_r2
```

```python
import math
print(math.sqrt(mean_squared_error(y_test, y_predict))) # root mean squared error
```

```python
print(mean_absolute_error(y_test, y_predict))
```

```python
root_mean_squared_error(y_test, y_predict)
```
